## User Defined Functions

```python
# Make a basic sinusoidal light curve

def make_time(Nobs,tspan ,tstart = 0, random = False):
    '''
    documentation will go here eventually
    
    required input:
            Nobs :      Number of observation point in curve (default 100)
            tspan :     The span of the lightcurve in days
    optional input:
            tstart :    The starting time of the light curve
            random :    True or False if the values are randomly set
    output:
            time :      nparray of time values
            
    '''

    # Make sure the number of observations is an integer
    Nobs = int(Nobs)

    tspan = float(tspan)
    tstart = float(tstart)
    tend = tstart + tspan
    
    if random:
        values = np.random.uniform(tstart,tend,Nobs)
        time = np.sort(values)
    else:
        time = np.linspace(tstart, tend, Nobs)
        
    return time
    


def make_flux(time, period, amp, mean = 0, phi = 0, noise_sigma = 0) :
    '''
    documentation will go here eventually

    required input:
            time :      nparry of time values
            period :    float or int value
            amp :       float or int value
    optional input:
            mean :      mean flux value (default is 0)
            phi :       phase shift of light curve (default 0)
    output:
            flux
    '''

    # convert values to float or int as needed
    period = float(period)
    amp = float(amp)
    
    Nobs = len(time)


    # create the flux array to match the time array
    flux = amp*np.sin(2*np.pi/period*time + phi)

    return flux
        
```
